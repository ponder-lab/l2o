```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import plot
```

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 6))
tests = {
    "choice_20": "Unroll=20, Depth=500",
    "choice_50": "Unroll=50, Depth=1250",
    "choice_100": "Unroll=100, Depth=2500",
    "choice_hard": "Unroll=20, Depth=500, Hard"
}


for test, name in tests.items():
    df = pd.read_csv(plot.get_test(test, data="summary"))
    axs.plot(df["meta_loss"], label=name)
axs.legend(bbox_to_anchor=(1, 1),)
axs.set_xlabel("Period (100 meta epochs; run 4x parallel)")
axs.set_ylabel("Log Meta Loss")
```
