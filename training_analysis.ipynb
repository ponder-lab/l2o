```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import plot
```

```python
def plot_training(tests, ax):
    for test in tests:
        df = pd.read_csv(plot.get_test(test, data="summary"))
        ax.plot(df["meta_loss"], label=plot.get_name(test))
    ax.set_xlabel("Period (100 meta epochs; run 4x parallel)")
    ax.set_ylabel("Meta Loss")
    ax.legend()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(16, 16))
plot_training(["choice_20", "choice_50", "choice_100", "choice_wider", "choice_bigger"], axs[0][0])
plot_training(["choice_hard", "choice_hard_50"], axs[0][1])
plot_training(["choice_mlp", "choice_mlp_deeper"], axs[1][0])
plot_training(["choice_20", "choice_sigmoid", "choice_leaky_relu"], axs[1][1])
plot_training(["choice_mlp", "choice_dbg", "choice_dbg2"], axs[2][0])
# plot_training(["scale_20", "rnnprop_20"], axs[2][1])
plot_training(["choice_nrscale", "choice_20"], axs[2][1])
fig.tight_layout()
```

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 6))
plot_training(["scale_20"], axs)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

for test, ax in zip(["rnnprop_20_adam", "rnnprop_mlp_adam"], axs):
    df = pd.read_csv(plot.get_test(test, data="summary"))
    ax2 = ax.twinx()
    ax.plot(df["meta_loss"], label="Meta Loss", color='C0')
    ax2.plot(df["imitation_loss"], label="Imitation Loss", color='C1')
    ax.set_title(plot.get_name(test))

    ax.set_xlabel("Period (100 meta epochs; run 4x parallel)")
    ax.set_ylabel("Meta Loss")
    ax.set_ylim(-0.045, 0.025)
    ax.legend()
    ax2.legend()
    ax2.set_ylabel("Imitation Loss")

fig.tight_layout()
```
