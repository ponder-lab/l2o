```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import plot
```

```python
def plot_eval(tests):
    fig, axs = plt.subplots(2, 1, figsize=(16, 6))
    plot.plot_stats_batch(tests, axs[0], end=1000, sma=1)
    axs[0].set_title("First 1000 Steps")
    plot.plot_stats_batch(tests, axs[1], end=0, sma=50)
    axs[1].set_title("All steps (50-step Moving Average)")
    fig.tight_layout()
```

```python
plot_eval(["adam", "choice-20x25C", "choice-20x50C", "choice-20x25C-hard"])
plot_eval(["adam", "choice-20x25M", "choice-20x50M", "choice-20x25M-hard"])
plot_eval(["adam", "rnnprop-20x25C", "rnnprop-20x25M", "rnnprop-20x25C-repeat"])
```
