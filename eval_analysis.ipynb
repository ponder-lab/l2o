```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import plot
```

```python
def plot_eval(tests):
    fig, axs = plt.subplots(2, 1, figsize=(16, 6))
    plot.plot_stats_batch(tests, axs, end=500, sma=1)
    axs[1].set_ylim(0.8, 1)
    fig.suptitle("First 500 Steps, Loss and Accuracy by Batch")

    fig, axs = plt.subplots(2, 1, figsize=(16, 6))
    plot.plot_stats_batch(tests, axs, end=0, sma=50)
    axs[1].set_ylim(0.95, 1)
    fig.suptitle("All steps (50-step Moving Average)")
```

```python
plot_eval(["adam", "rmsprop", "choice-20x25M", "choice-20x25C"])
```

```python
plot_eval(["adam", "rmsprop", "rnnprop-20x25M", "rnnprop-20x25C"])
```

```python
plot_eval(["adam", "rmsprop", "rnnprop-20x25CA", "rnnprop-20x25CR", "rnnprop-20x25CAR"])
```

```python
tests = ["adam", "rmsprop", "rnnprop-20x25C"]
fig, axs = plt.subplots(2, 1, figsize=(16, 6))
plot.plot_stats_batch(tests, axs, end=500, sma=1, period=30)
axs[1].set_ylim(0.8, 1)
fig.suptitle("First 500 Steps, Loss and Accuracy by Batch")

fig, axs = plt.subplots(2, 1, figsize=(16, 6))
plot.plot_stats_batch(tests, axs, end=0, sma=50, period=30)
axs[1].set_ylim(0.98, 1)
fig.suptitle("All steps (50-step Moving Average)")
```
