```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import plot
```

```python
fig, axs = plt.subplots(1, 1, figsize=(16, 12))
plot.plot_phase(["choice_20", "choice_50", "choice_100", "adam", "rmsprop"], axs)
axs.set_xlim([0.7, 1.0])
axs.set_ylim([0.7, 1.0])
axs.legend()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
plot.plot_stats(["choice_20", "choice_50", "choice_100", "adam", "rmsprop"], axs)
```

```python
KEYS = [
    "loss", "val_loss",
    "sparse_categorical_accuracy",
    "val_sparse_categorical_accuracy"
]
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
for e in [19, 39, 59, 79, 99]:
    d = np.load("results/choice_20/period_{}/conv_train.npz".format(e))
    for ax, val in zip([*axs[0], *axs[1]], KEYS):
        y_val = np.log(d[val]) if val.endswith("loss") else d[val]
        plot.plot_band(ax, np.arange(25), y_val, label=e)

for ax in [*axs[0], *axs[1]]:
    ax.set_xticks(np.arange(25))
    ax.set_xlabel("Epoch")

axs[1][1].legend()

axs[0][0].set_ylabel("Log Training Loss")
axs[0][1].set_ylabel("Log Validation Loss")
axs[1][0].set_ylabel("Training Accuracy")
axs[1][1].set_ylabel("Validation Accuracy")
```

```python
fig, axs = plt.subplots(1, 1, figsize=(16, 12))
COLOR_CYCLE = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

for e, c in zip([19, 39, 59, 79, 99], COLOR_CYCLE):
    d = np.load("results/choice_20/period_{}/conv_train.npz".format(e))
    x = np.mean(d["sparse_categorical_accuracy"], axis=0)
    y = np.mean(d["val_sparse_categorical_accuracy"], axis=0)
    axs.quiver(
        x[:-1], y[:-1], x[1:] - x[:-1], y[1:] - y[:-1],
        width=0.002, headwidth=5, headlength=5, color=c,
        scale_units="xy", angles="xy", scale=1, label=e)
axs.legend()
```
